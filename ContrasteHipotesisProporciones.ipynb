## Introducción

En este documento, realizaremos un contraste de hipótesis para comparar las proporciones de dos grupos en una variable dicotómica (por ejemplo, la proporción de éxito en un examen). Utilizaremos el lenguaje R para realizar el análisis.

## Paso 1: Definir el Problema

Supongamos que tenemos dos grupos de estudiantes que han tomado un examen:
- Grupo 1: 30 aprobados de 50 estudiantes.
- Grupo 2: 45 aprobados de 80 estudiantes.

Queremos probar si la proporción de aprobados es diferente entre los dos grupos.

## Paso 2: Plantear las Hipótesis

- **Hipótesis nula ($H_0$)**: Las proporciones de éxito son iguales.
  
  $$ H_0: p_1 = p_2 $$
  
- **Hipótesis alternativa ($H_a$)**: Las proporciones de éxito son diferentes.
  
  $$ H_a: p_1 \neq p_2 $$

## Paso 3: Definir el Nivel de Significancia

Establecemos el nivel de significancia $\alpha = 0.05$.

## Paso 4: Recopilar los Datos

Los datos de los dos grupos son los siguientes:

- Grupo 1:
  - Aprobados: $x_1 = 30$
  - Muestra total: $n_1 = 50$

- Grupo 2:
  - Aprobados: $x_2 = 45$
  - Muestra total: $n_2 = 80$

## Paso 5: Calcular las Proporciones

Calculamos las proporciones de aprobados para cada grupo:

$$
\hat{p}_1 = \frac{x_1}{n_1} = \frac{30}{50} = 0.6
$$

$$
\hat{p}_2 = \frac{x_2}{n_2} = \frac{45}{80} = 0.5625
$$

## Paso 6: Calcular la Proporción Combinada

Calculamos la proporción combinada:

$$
\hat{p} = \frac{x_1 + x_2}{n_1 + n_2} = \frac{30 + 45}{50 + 80} = \frac{75}{130} \approx 0.5769
$$

## Paso 7: Calcular el Estadístico de Prueba

El estadístico de prueba $z$ se calcula como:

$$
z = \frac{\hat{p}_1 - \hat{p}_2}{\sqrt{\hat{p} (1 - \hat{p}) \left(\frac{1}{n_1} + \frac{1}{n_2}\right)}}
$$

Sustituyendo los valores:

- $1 - \hat{p} \approx 0.4231$

$$
z = \frac{0.6 - 0.5625}{\sqrt{0.5769 \cdot 0.4231 \left(\frac{1}{50} + \frac{1}{80}\right)}}
$$

Sustituyamos los valores en la fórmula.

1. Calcular $1 - \hat{p}$:
   $$
   1 - \hat{p} \approx 1 - 0.5769 \approx 0.4231
   $$

2. Calcular el denominador:
   $$
   \sqrt{\hat{p} (1 - \hat{p}) \left(\frac{1}{n_1} + \frac{1}{n_2}\right)} = \sqrt{0.5769 \times 0.4231 \left(\frac{1}{50} + \frac{1}{80}\right)}
   $$

   Primero calculamos $\frac{1}{50} + \frac{1}{80}$:
   $$
   \frac{1}{50} = 0.02, \quad \frac{1}{80} = 0.0125 \quad \Rightarrow \quad \frac{1}{50} + \frac{1}{80} = 0.02 + 0.0125 = 0.0325
   $$

   Ahora sumamos:
   $$
   \hat{p} (1 - \hat{p}) \left(\frac{1}{n_1} + \frac{1}{n_2}\right) \approx 0.5769 \times 0.4231 \times 0.0325
   $$

# EN PYTHON

Suposiciones del Problema

- Tratamiento A: Se aplicó a 50 pacientes, de los cuales 30 se curaron.
- Tratamiento B: Se aplicó a 50 pacientes, de los cuales 20 se curaron.

Queremos probar si hay una diferencia significativa en las proporciones de pacientes curados entre los dos tratamientos.

Código Completo en Python

```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats  # Cambiamos esto para importar 'stats' de scipy.

# Paso 1: Definir los datos
n_A = 50  # Tamaño del grupo A
x_A = 30  # Pacientes curados en el grupo A

n_B = 50  # Tamaño del grupo B
x_B = 20  # Pacientes curados en el grupo B

# Paso 2: Crear un DataFrame para visualizar los datos
data = pd.DataFrame({
    'Tratamiento': ['A'] * n_A + ['B'] * n_B,
    'Curado': [1] * x_A + [0] * (n_A - x_A) + [1] * x_B + [0] * (n_B - x_B)
})



# Paso 3: Calcular las proporciones
p_A = x_A / n_A
p_B = x_B / n_B


# Paso 4: Generar gráfico de barras para las proporciones
plt.bar(['Tratamiento A', 'Tratamiento B'], [p_A, p_B], color=['blue', 'orange'])
plt.title('Proporciones de Pacientes Curados por Tratamiento')
plt.ylabel('Proporción de Curados')
plt.ylim(0, 1)  # Para que las proporciones vayan de 0 a 1
plt.axhline(y=p_A, color='blue', linestyle='--', label=f'Curados A: {p_A:.2f}')
plt.axhline(y=p_B, color='orange', linestyle='--', label=f'Curados B: {p_B:.2f}')
plt.legend()
plt.show()

```

```python
# Paso 5: Realizar el contraste de hipótesis para dos proporciones
# Proporción combinada
p_pool = (x_A + x_B) / (n_A + n_B)

# Estadístico z
z = (p_A - p_B) / np.sqrt(p_pool * (1 - p_pool) * (1/n_A + 1/n_B))

# Calcular el valor p (bilateral) usando scipy.stats
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Paso 6: Definir el nivel de significancia
alpha = 0.05

# Paso 7: Tomar decisión
print(f"\nEstadístico z: {z:.4f}")
print(f"Valor p: {p_value:.4f}")

if p_value < alpha:
    print("Rechazar la hipótesis nula. Hay una diferencia significativa entre los tratamientos.")
else:
    print("No se puede rechazar la hipótesis nula. No hay suficiente evidencia para afirmar que hay una diferencia significativa.")
```

# Otro ejemplo con distintas librerías estadísticas

```python
# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from tabulate import tabulate  # Librería para mejorar la presentación de tablas

# Paso 1: Definir los datos
n_A = 50  # Tamaño del grupo A
x_A = 30  # Pacientes curados en el grupo A

n_B = 50  # Tamaño del grupo B
x_B = 20  # Pacientes curados en el grupo B

# Paso 2: Crear una tabla de contingencia
data = pd.DataFrame({
    'Curados': [x_A, x_B],
    'No_Curados': [n_A - x_A, n_B - x_B]
}, index=['Tratamiento A', 'Tratamiento B'])

print("Tabla de Contingencia:")
print(tabulate(data, headers='keys', tablefmt='grid'))



# Paso 6: Generar un gráfico de barras para las proporciones
proporciones = data.div(data.sum(axis=1), axis=0)

proporciones.plot(kind='bar', color=['blue', 'orange'])
plt.title('Proporciones de Pacientes Curados por Tratamiento')
plt.ylabel('Proporción de Curados')
plt.ylim(0, 1)  # Para que las proporciones vayan de 0 a 1
plt.axhline(y=proporciones.loc['Tratamiento A', 'Curados'], color='blue', linestyle='--', label=f'Curados A: {proporciones.loc["Tratamiento A", "Curados"]:.2f}')
plt.axhline(y=proporciones.loc['Tratamiento B', 'Curados'], color='orange', linestyle='--', label=f'Curados B: {proporciones.loc["Tratamiento B", "Curados"]:.2f}')
plt.legend()
plt.show()
```

```python
# Paso 3: Realizar el contraste de hipótesis para dos proporciones usando statsmodels
# Crear un arreglo de conteo para el análisis
successes = np.array([x_A, x_B])
trials = np.array([n_A, n_B])

# Realizar la prueba de z-test de proporciones
z_stat, p_value = sm.stats.proportions_ztest(successes, trials)

# Paso 4: Definir el nivel de significancia
alpha = 0.05

# Paso 5: Calcular el poder estadístico
# Calcular la proporción de éxito para cada grupo
p_A = successes[0] / trials[0]
p_B = successes[1] / trials[1]

# Proporción combinada
p_pool = (successes[0] + successes[1]) / (trials[0] + trials[1])

# Diferencia entre proporciones observada
effect_size = p_A - p_B

# Calcular el poder de la prueba
# Define la relación entre los tamaños de los grupos
ratio = n_A / n_B
power = sm.stats.NormalIndPower().power(effect_size=effect_size / np.sqrt(p_pool * (1 - p_pool)), nobs1=n_A, ratio=ratio, alpha=alpha)

# Paso 6: Tomar decisión
print(f"\nEstadístico z: {z_stat:.4f}")
print(f"Valor p: {p_value:.4f}")
print(f"Poder estadístico: {power:.4f}")

if p_value < alpha:
    print("Rechazar la hipótesis nula. Hay una diferencia significativa entre los tratamientos.")
else:
    print("No se puede rechazar la hipótesis nula. No hay suficiente evidencia para afirmar que hay una diferencia significativa.")
```

El poder estadístico es un concepto clave en la estadística que se refiere a la capacidad de una prueba de hipótesis para detectar un efecto, si es que realmente existe. En este documento, exploraremos el significado del poder estadístico, cómo se calcula y por qué es importante en el contexto de la comparación de proporciones.

## Definición de Poder Estadístico

El poder estadístico de una prueba es la probabilidad de rechazar la hipótesis nula $H_0$ cuando es falsa. En otras palabras, el poder indica la capacidad de una prueba para detectar diferencias reales o efectos en los datos.

$$
\text{Poder} = P(\text{Rechazar } H_0 | H_0 \text{ es falsa})
$$

### Factores que Afectan el Poder Estadístico

Varios factores influyen en el poder estadístico de una prueba:

1. **Tamaño del Efecto**: Cuanto mayor sea la diferencia entre los grupos (por ejemplo, entre proporciones), mayor será el poder de la prueba. El tamaño del efecto puede ser medido por la diferencia de medias o proporciones.

2. **Tamaño de la Muestra**: A medida que aumenta el tamaño de la muestra, se incrementa el poder estadístico. Esto se debe a que una muestra más grande proporciona más información sobre la población y reduce la variabilidad.

3. **Nivel de Significancia $\alpha$**: Este es el umbral que se utiliza para decidir si se rechaza $H_0$. Si se establece un nivel de significancia más alto, el poder aumenta, ya que es más fácil rechazar $H_0$. Sin embargo, esto también aumenta la probabilidad de cometer un error Tipo I.

4. **Variabilidad en los Datos**: Una menor variabilidad en los datos (menor desviación estándar) aumenta el poder de la prueba, ya que las diferencias entre grupos se destacan más.
