# El análisis bivariante
es una técnica estadística utilizada para explorar la relación entre dos variables. Se utiliza para entender cómo una variable se comporta en función de otra y permite identificar patrones, correlaciones y posibles causas o efectos. Este tipo de análisis es fundamental en la investigación, ya que ayuda a establecer vínculos y comprender interacciones dentro de los datos.

Componentes del Análisis Bivariante:
Variables:

**Variable independiente** (X): Es la variable que se supone que influye o predice en la otra variable.
Variable dependiente (Y): Es la variable que se mide o se espera que cambie en respuesta a la variable independiente.

**Métodos de Análisis:**
Dependiendo del tipo de variables (categóricas o numéricas), se puede utilizar diferentes técnicas:

Variables numéricas:

Correlación de Pearson: Mide la fuerza y la dirección de la relación lineal entre dos variables.\
Regresión lineal: Permite modelar la relación entre las dos variables para hacer predicciones.\
Diagrama de dispersión: Visualiza la relación entre las dos variables numéricas.
Una variable categórica y una numérica:

**Pruebas de hipótesis (como la prueba t):**
Se utiliza para comparar medias entre dos grupos (por ejemplo, calificaciones de hombres versus mujeres).


**Tablas de contingencia:** Muestran la frecuencia de diferentes combinaciones de categorías.

Chi-cuadrado: Prueba para determinar si existe una asociación significativa entre las dos variables categóricas.

**Objetivos del Análisis Bivariante:**



1.   Identificar relaciones: Determinar si hay una relación entre las dos variables y su naturaleza (positiva, negativa o ninguna).
2.   Evaluar la fuerza de la relación: Medir cuán fuerte es la relación mediante coeficientes o pruebas estadísticas.
3. Realizar inferencias: Hacer pronósticos o inferencias basadas en la relación observada.
4. Visualizar la relación: Utilizar gráficos para facilitar la comprensión de cómo las variables interactúan entre sí.

**Conclusión:**
El análisis bivariante es una herramienta poderosa en estadística que permite a los investigadores y analistas entender mejor las interacciones entre dos variables. Es esencial en muchos campos, como la economía, la psicología, la biología y las ciencias sociales, ya que ayuda a fundamentar la toma de decisiones y a identificar áreas para un análisis más profundo.

# Correlación Lineal y Regresión Lineal

## Definición de Correlación Lineal

La correlación lineal mide la relación entre dos variables. Existen varios coeficientes utilizados para medir la correlación, destacando los de Pearson y Spearman.

### Correlación de Pearson

La correlación de Pearson, denotada como $r$, se calcula como:

$$
r = \frac{Cov(X, Y)}{\sigma_X \sigma_Y}
$$

donde:
- $Cov(X, Y)$ es la covarianza entre las variables $X$ y $Y$.
- $\sigma_X$ y $\sigma_Y$ son las desviaciones estándar de $X$ y $Y$, respectivamente.

El valor de $r$ oscila entre -1 y 1:
- $r = 1$: correlación positiva perfecta.
- $r = -1$: correlación negativa perfecta.
- $r = 0$: sin correlación.

### Correlación de Spearman

La correlación de Spearman, denotada como $\rho$, se basa en los rangos de los datos en lugar de sus valores absolutos. Se define como:

$$
\rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)}
$$

donde:
- $d_i$ es la diferencia entre los rangos de cada par de datos.
- $n$ es el número de observaciones.

Al igual que la de Pearson, $\rho$ también varía entre -1 y 1.

## Diferencias respecto a la Regresión Lineal

Aunque tanto la correlación como la regresión analizan la relación entre variables, existen diferencias clave:

1. **Propósito**:
   - **Correlación**: mide la fuerza y dirección de una relación lineal entre dos variables.
   - **Regresión**: modela la relación entre variables, permitiendo predecir valores de una variable (dependiente) a partir de otra (independiente).

2. **Relación**:
   - **Correlación**: no implica causalidad; solo indica si las variables están relacionadas.
   - **Regresión**: busca establecer una relación causal entre variables.

3. **Ecuación**:
   - **Correlación**: no se expresa en forma de ecuación, sino como un coeficiente.
   - **Regresión**: se expresa en forma de ecuación lineal, por ejemplo:
   $$
   Y = a + bX
   $$
   donde $Y$ es la variable dependiente, $X$ es la variable independiente, $a$ es la intersección y $b$ es la pendiente.

## Conclusión

La correlación y la regresión son herramientas estadísticas fundamentales, cada una con su propósito específico. Es importante elegir el método adecuado según el análisis que se desea llevar a cabo.

Volvamos a la base de datos IRIS para calcular el test de normalidad, también con lenguaje natural.

```python
# Paso 1: Importar librerías

import seaborn as sns
from scipy import stats

# Paso 2: Cargar la base de datos Iris
iris = sns.load_dataset('iris')

# Paso 3: Realizar el test de normalidad de Shapiro-Wilk
normality_results = {}
for column in iris.select_dtypes(include=['float64']):
    stat, p_value = stats.shapiro(iris[column])
    normality_results[column] = (stat, p_value)

# Mostrar resultados
for variable, (statistic, p_value) in normality_results.items():
    print(f'Variable: {variable}')
    print(f'Estadístico de Shapiro: {statistic:.4f}, Valor p: {p_value:.4f}')
    if p_value > 0.05:
        print("  - La variable se distribuye normalmente (no se rechaza la hipótesis nula).\n")
    else:
        print("  - La variable no se distribuye normalmente (se rechaza la hipótesis nula).\n")
```
