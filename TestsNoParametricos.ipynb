## Introducción

Las pruebas no paramétricas son técnicas estadísticas que se utilizan cuando los datos no cumplen con los supuestos necesarios para aplicar pruebas paramétricas, tales como la normalidad de la distribución y la homogeneidad de la varianza. Estas pruebas son especialmente útiles cuando se trabaja con muestras pequeñas o datos que no se distribuyen normalmente.

## Cuándo utilizar pruebas no paramétricas

Se recomienda utilizar pruebas no paramétricas en las siguientes situaciones:

1. **Datos no distribuidos normalmente**: Cuando los datos no siguen una distribución normal, las pruebas paramétricas, como la prueba *t* de Student, pueden no ser adecuadas.

2. **Muestra pequeña**: Cuando el tamaño de la muestra es pequeño (por lo general, menos de 30 observaciones), la verificación de la normalidad puede ser difícil, y las pruebas no paramétricas son más robustas.

3. **Escalas de medición restringidas**: Cuando los datos son de orden o no se pueden asumir intervalos iguales, como en escalas ordinales.

4. **Presencia de valores atípicos**: Los valores atípicos pueden influir en las pruebas paramétricas, y las pruebas no paramétricas son menos sensibles a estos.

## Pruebas no paramétricas para la comparación de medias

### 1. Prueba de Wilcoxon (Mann-Whitney U)

La prueba de Wilcoxon, también conocida como la prueba de Mann-Whitney U, se utiliza para comparar las medianas de dos grupos independientes.

**Hipótesis**:
  - Hipótesis nula $H_0$: Las dos muestras provienen de la misma población.
  - Hipótesis alternativa $H_1\$: Las muestras provienen de poblaciones diferentes.

**Estadístico de prueba**:
La estadística $(U)$ es calculada a partir de las clasificaciones de las observaciones en ambas muestras. La fórmula general para \(U\) es:

$$
U = R_1 - \frac{n_1(n_1 + 1)}{2}
$$

donde:
- $R_1$ es la suma de los rangos del primer grupo.
- $n_1$ es el tamaño del primer grupo.

### 2. Prueba de Kruskal-Wallis

La prueba de Kruskal-Wallis es una extensión de la prueba de Wilcoxon y se utiliza para comparar las medianas de tres o más grupos independientes.

**Hipótesis**:
  - Hipótesis nula $H_0$: Las muestras provienen de la misma población.
  - Hipótesis alternativa $H_1$: Al menos una muestra proviene de una población diferente.

**Estadístico de prueba**:
El estadístico \(H\) se calcula como sigue:

$$
H = \frac{12}{N(N+1)} \sum \frac{R_j^2}{n_j} - 3(N + 1)
$$

donde:
- $N$ es el número total de observaciones.
- $R_j$ es la suma de rangos para el grupo $j$.
- $n_j$ es el número de observaciones en el grupo $j$.

Kruskal, W. H., & Wallis, W. A. (1952). Use of ranks in one-criterion variance analysis. Journal of the American Statistical Association, 47(260), 583-621. https://doi.org/10.1080/01621459.1952.10483441

### 3. Prueba de Friedman

La prueba de Friedman se utiliza para comparar las medias de tres o más grupos relacionados. Es una alternativa no paramétrica a la ANOVA de medidas repetidas.

**Hipótesis**:
  - Hipótesis nula $H_0$: No hay diferencias en las medias de los grupos.
  - Hipótesis alternativa $H_1$: Al menos una media es diferente.

**Estadístico de prueba**:
Se utiliza el estadístico \(Q\):

$$
Q = \frac{12}{nk(k + 1)} \sum R_j^2 - 3n(k + 1)
$$

donde:
- \(n\) es el número de bloques (sujetos).
- \(k\) es el número de tratamientos (grupos).

Friedman, M. (1937). The use of ranks to avoid the assumption of normality implicit in the analysis of variance. Journal of the American Statistical Association, 32(200), 675-701. https://doi.org/10.1080/01621459.1937.10503522

## Conclusión

Las pruebas no paramétricas son herramientas valiosas en estadística, especialmente cuando los datos no cumplen con los supuestos de las pruebas paramétricas. Al ser menos sensibles a las características de los datos, proporcionan una alternativa robusta para la comparación de medias en investigaciones donde la normalidad y la homogeneidad de varianza no pueden asumirse.


## La Correlación lineal Spearman (No paramétrico)

La correlación de Spearman es una medida no paramétrica que evalúa la relación monotónica entre dos variables. A diferencia de la correlación de Pearson, que asume una relación lineal y requiere que ambas variables sean distribuidas normalmente, la correlación de Spearman puede ser utilizada con datos ordinales o cuando las suposiciones de normalidad no son cumplidas.

## ¿Cuándo utilizar la correlación de Spearman?

Se recomienda utilizar la correlación de Spearman en las siguientes situaciones:

1. **Datos ordinales**: Cuando los datos se recopilan en escalas ordinales, es decir, donde se puede establecer un orden, pero no se puede asumir igual distancia entre las categorías.

2. **No linealidad**: Cuando se sospecha que la relación entre las variables no es lineal, pero sí monotónica (es decir, que a medida que una variable aumenta, la otra también lo hace o disminuye).

3. **Datos no normales**: Cuando los datos no se distribuyen normalmente, lo que hace que la correlación de Pearson no sea adecuada.

4. **Datos con valores atípicos**: La correlación de Spearman es menos sensible a los valores atípicos que la correlación de Pearson.

## Cálculo de la Correlación de Spearman

La correlación de Spearman se calcula en los siguientes pasos:

1. **Clasificar los datos**: Asigna rangos a las observaciones. Si hay empates, se utiliza el promedio de los rangos para esos valores.

2. **Calcular la diferencia de rangos**: Para cada par de observaciones, calcula la diferencia de los rangos:

$$
d_i = R_{X_i} - R_{Y_i}
$$

donde:
- $R_{X_i}$ es el rango de la variable $X$.
- $R_{Y_i}$ es el rango de la variable $Y$.

3. **Calcular la suma de los cuadrados de las diferencias**:

$$
\sum d_i^2
$$

4. **Aplicar la fórmula de Spearman**:

La fórmula para calcular el coeficiente de correlación de Spearman $r_s$ es:

$$
r_s = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)}
$$

donde:
- $n$ es el número de pares de datos.

## Interpretación del Coeficiente de Correlación de Spearman

- $r_s = 1$: Relación perfecta positiva.
- $r_s = -1$: Relación perfecta negativa.
- $r_s = 0$: No hay relación monotónica entre las variables.

Los valores de \(r_s\) entre 0 y 1 indican la fuerza de la relación, donde valores más cerca de 1 o -1 indican una relación más fuerte.

## EJEMPLO PASO A PASO

Creamos una muestra con un modelo exponencial y dibujamos el histograma de ambos grupos



```python
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Fijar una semilla para la reproducibilidad
np.random.seed(42)

# Simular dos grupos de datos de tamaño n=30
n = 30
grupo1 = np.random.exponential(scale=1, size=n)  # Distribución exponencial (no normal)
grupo2 = np.random.exponential(scale=2, size=n)  # Distribución exponencial (no normal)


# Visualizar histogramas y curvas normales
plt.figure(figsize=(12, 6))

# Histograma para el Grupo 1
sns.histplot(grupo1, bins=15, kde=True, color='blue', label='Grupo 1', stat='density', alpha=0.6)

# Histograma para el Grupo 2
sns.histplot(grupo2, bins=15, kde=True, color='orange', label='Grupo 2', stat='density', alpha=0.6)

# Ajustar y graficar la curva normal para el Grupo 1
mu1, std1 = stats.norm.fit(grupo1)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p1 = stats.norm.pdf(x, mu1, std1)
#plt.plot(x, p1, 'b--', linewidth=2, label='Curva Normal Grupo 1')

# Ajustar y graficar la curva normal para el Grupo 2
mu2, std2 = stats.norm.fit(grupo2)
p2 = stats.norm.pdf(x, mu2, std2)
#plt.plot(x, p2, 'orange', linestyle='--', linewidth=2, label='Curva Normal Grupo 2')

plt.title('Histogramas y Curvas Normales de los Grupos')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

### Testeamos la normalidad de las muestras

```python
# Verificar la normalidad utilizando el test de Shapiro-Wilk
stat1, p1 = stats.shapiro(grupo1)
stat2, p2 = stats.shapiro(grupo2)

print(f"Grupo 1: Estadístico = {stat1:.4f}, p-valor = {p1:.4f} (p-value)")
print(f"Grupo 2: Estadístico = {stat2:.4f}, p-valor = {p2:.4f} (p-value)")

# Evaluar si los grupos son normales o no
if p1 > 0.05:
    print("Grupo 1: La serie proviene de una población con distribución normal.")
else:
    print("Grupo 1: La serie NO proviene de una población con distribución normal.")

if p2 > 0.05:
    print("Grupo 2: La serie proviene de una población con distribución normal.")
else:
    print("Grupo 2: La serie NO proviene de una población con distribución normal.")
```

### Supuestos de las Pruebas Paramétricas

Las pruebas paramétricas, como la t de Student, requieren que se cumplan ciertos supuestos, incluyendo:

1. **Normalidad**: Las muestras deben seguir una distribución normal.
2. **Homogeneidad de Varianzas**: Las varianzas de los grupos comparados deben ser iguales (esto es lo que evalúa el test de Levene).

Cuando los datos no son normales, los supuestos correspondientes a las pruebas paramétricas ya no son válidos. En este caso, se desaconseja el uso de pruebas paramétricas, independientemente de si las varianzas son homogéneas o no.

### Robustez de las Pruebas No Paramétricas

Las pruebas no paramétricas son menos sensibles a los supuestos de distribución y no requieren que los datos sean normales ni que las varianzas sean homogéneas. Por consiguiente, cuando se ha confirmado que los datos no siguen una distribución normal, se puede proceder directamente a aplicar un test no paramétrico:

- **Mann-Whitney U**: Este test compara las medianas de dos grupos independientes sin asumir normalidad o homogeneidad de varianzas.
- **Kruskal-Wallis**: Este test se utiliza para comparar más de dos grupos independientes.

```python
#Realizar el test de Mann-Whitney U
stat_uw, p_uw = stats.mannwhitneyu(grupo1, grupo2)

# Imprimir los resultados de la prueba de Mann-Whitney U
print(f"Prueba de Mann-Whitney U: Estadístico = {stat_uw}, p-valor = {p_uw}")

# Paso 6: Mostrar resultados sobre la hipótesis nula
if p_uw < 0.05:
    print("Se rechaza la hipótesis nula: hay diferencias significativas entre los grupos.")
else:
    print("No se rechaza la hipótesis nula: no hay diferencias significativas entre los grupos.")
```

El test de Mann-Whitney U es una prueba no paramétrica que se utiliza para comparar las medianas de dos grupos independientes. Este test es especialmente útil cuando los datos no cumplen con los supuestos de normalidad requeridos por las pruebas paramétricas. Al realizar este test, es fundamental entender el proceso de aceptación de la hipótesis alternativa.

## Hipótesis del Test de Mann-Whitney U

En el contexto del test de Mann-Whitney U, se definen las siguientes hipótesis:

- **Hipótesis nula ($H_0$)**: No hay diferencias en las medianas de los dos grupos. Esto significa que las dos muestras provienen de la misma distribución.
  
- **Hipótesis alternativa ($H_1$)**: Hay diferencias en las medianas de los dos grupos. Esto implica que al menos una de las muestras proviene de una población diferente de la otra.

## Criterios para la Aceptación de la Hipótesis Alternativa

La aceptación de la hipótesis alternativa en el test de Mann-Whitney U se basa en el valor p resultante de la prueba. Los criterios son los siguientes:

1. **Umbral del p-valor**: Se establece un nivel de significancia ($\alpha$), generalmente 0.05 (5%). Este umbral indica la probabilidad de rechazar la hipótesis nula cuando es verdadera.

2. **Interpretación del p-valor**:
   - Si el p-valor calculado es **menor o igual a $\alpha$** (p ≤ 0.05), se rechaza la hipótesis nula ($H_0$) y se acepta la hipótesis alternativa ($H_1$).
   - Si el p-valor es **mayor que $\alpha$** (p > 0.05), no se tiene suficiente evidencia para rechazar la hipótesis nula, y se concluye que no hay diferencias significativas entre los grupos.

# CARGAMOS LA BASE DE DATOS

```python
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el conjunto de datos de diabetes
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = [
    'Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
    'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome'
]
diabetes_data = pd.read_csv(url, names=columns)

# Mostrar las primeras filas del conjunto de datos
diabetes_data.head()

```

## ANALIZAMOS UNA VARIABLE BMI

Para realizar un análisis de comparación de medias utilizando un test no paramétrico y graficar un boxplot con Seaborn en Python, primero seleccionaremos una variable numérica y la agruparemos con una variable categórica. En la base de datos Pima Indian Diabetes, una buena opción es comparar el Índice de Masa Corporal (BMI) entre los grupos definidos por la variable Outcome (diabetes o no diabetes).

```python
# Seleccionar la variable numérica (BMI) y la variable categórica (Outcome)
bmi_values = diabetes_data['BMI']
outcome_values = diabetes_data['Outcome']

# Realizar el test de Mann-Whitney U (no paramétrico)
group0 = bmi_values[outcome_values == 0]  # Grupo sin diabetes
group1 = bmi_values[outcome_values == 1]  # Grupo con diabetes


# Graficar el boxplot
sns.boxplot(x='Outcome', y='BMI', data=diabetes_data)
plt.title('Comparación de BMI por Resultado de Diabetes')
plt.xlabel('Diabetes (0 = No, 1 = Sí)')
plt.ylabel('Índice de Masa Corporal (BMI)')
plt.show()
```

Para analizar la normalidad de la variable Índice de Masa Corporal (BMI) en los dos grupos (personas con diabetes y personas sin diabetes), se puede utilizar el test de normalidad de Shapiro-Wilk. Este test es comúnmente utilizado para evaluar la normalidad de una distribución en muestras pequeñas y es adecuado para nuestra comparación.

El test de Shapiro-Wilk verifica la hipótesis nula de que los datos provienen de una distribución normal. Si el valor p es menor que un nivel de significancia predefinido (usualmente 0.05), se rechaza la hipótesis nula, sugiriendo que los datos no siguen una distribución normal.

```python
from scipy.stats import shapiro

# Realizar el test de Shapiro-Wilk para cada grupo
stat0, p0 = shapiro(group0)
stat1, p1 = shapiro(group1)

# Mostrar resultados
print(f'Grupo 0 (Sin Diabetes): Estadística W = {stat0}, valor-p = {p0}')
print(f'Grupo 1 (Con Diabetes): Estadística W = {stat1}, valor-p = {p1}')
```

3. Interpretación de los Resultados

Si el valor p para el grupo sin diabetes (Outcome = 0) es menor que 0.05, podemos concluir que los BMI en este grupo no siguen una distribución normal.
Si el valor p para el grupo con diabetes (Outcome = 1) es menor que 0.05, indicaría lo mismo para ese grupo.

4. Argumentación Validación del Test No Paramétrico:

Si los resultados del test de Shapiro-Wilk muestran que al menos uno de los grupos no es normal, esto justifica el uso del test no paramétrico (como el test de Mann-Whitney U) para comparar las medias entre los dos grupos.
Distribución Normal: En caso de que ambos grupos muestren valores p superiores a 0.05, podríamos considerar usar un test paramétrico como el test t de Student, que requiere que los datos sean aproximadamente normales.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Crear un histograma con Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=diabetes_data, x='BMI', hue='Outcome', multiple='stack', kde=True, bins=30)

# Añadir etiquetas y título
plt.title('Distribución del Índice de Masa Corporal (BMI) por Resultado de Diabetes')
plt.xlabel('Índice de Masa Corporal (BMI)')
plt.ylabel('Frecuencia')
plt.legend(title='Diabetes (0 = No, 1 = Sí)')
plt.grid(True)
plt.show()
```

El test de Mann-Whitney U es un test no paramétrico utilizado para comparar dos grupos independientes cuando se desea evaluar si hay diferencias en sus distribuciones. Al argumentar sobre el uso del test de Mann-Whitney U para analizar el Índice de Masa Corporal (BMI) entre personas con diabetes y personas sin diabetes, se pueden considerar los siguientes puntos:

1. Naturaleza No Paramétrica

Distribuciones No Normales: El test de Mann-Whitney U no requiere que las muestras sigan una distribución normal, lo que lo hace adecuado cuando el test de normalidad (como Shapiro-Wilk) indica que al menos uno de los grupos no tiene una distribución normal. Esto es fundamental, ya que si se utilizan tests paramétricos (como el test t de Student) en datos no normales, los resultados pueden ser engañosos o incorrectos.

2. Comparación de Medianas

Enfoque en Medianas: Este test evalúa la hipótesis nula de que las dos muestras provienen de la misma población en términos de sus distribuciones mediales. Por lo tanto, se puede usar para determinar si hay una diferencia significativa en las medianas de BMI entre los grupos con y sin diabetes. Esto es especialmente relevante en el análisis de biomarcadores o métricas como el BMI.

3. Robustez ante Valores Atípicos

Menor Sensibilidad: A diferencia de los tests paramétricos, el test de Mann-Whitney U es menos sensible a los valores atípicos ya que se basa en rangos en lugar de medias. Esto es beneficioso en conjuntos de datos de salud, donde los valores atípicos son comunes.

4. Simplicidad y Facilidad de Ejecución

Facilidad de Interpretación: La ejecución y la interpretación del test de Mann-Whitney U son generalmente más sencillas, lo que permite una rápida comprensión de las diferencias entre grupos sin la necesidad de cumplir estrictos supuestos como la homogeneidad de varianzas.

5. Grupos Independientes

Diseño de Grupos: El test es útil cuando se comparan dos grupos independientes, tal como en este caso, donde se analizan las diferentes condiciones de salud (con diabetes frente a sin diabetes).

6. Aplicación en Estudios de Salud

Relevancia en Medicina: En el campo de la epidemiología y la salud pública, el uso de tests no paramétricos es común, especialmente cuando las variables de interés (como el BMI) pueden no cumplir con los supuestos necesarios para los tests paramétricos.

```python
from scipy.stats import mannwhitneyu

group0 = diabetes_data['BMI'][diabetes_data['Outcome'] == 0]  # Sin diabetes
group1 = diabetes_data['BMI'][diabetes_data['Outcome'] == 1]  # Con diabetes

# Paso 3: Realizar el test de Mann-Whitney U
stat, p_value = mannwhitneyu(group0, group1, alternative='two-sided')  # Alternativa 'dos lados'

# Paso 4: Mostrar los resultados
print(f'Estadística U: {stat}, valor-p: {p_value}')

# Paso 5: Interpretar los resultados
alpha = 0.05
if p_value < alpha:
    print("Se rechaza la hipótesis nula: hay diferencias significativas en BMI entre los dos grupos.")
else:
    print("No se rechaza la hipótesis nula: no hay diferencias significativas en BMI entre los dos grupos.")
```

Resultados:

Se imprime la estadística U y el valor p obtenidos a partir del test.
Se establece un nivel de significancia (alpha) de 0.05 y se interpreta el resultado: si el valor p es menor que alpha, se rechaza la hipótesis nula, lo que sugiere que hay diferencias significativas en el BMI entre los grupos.
