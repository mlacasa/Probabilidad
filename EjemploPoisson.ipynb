# Distribución de Poisson

La distribución de Poisson es una distribución de probabilidad **discreta** que se utiliza para modelar el número de eventos que ocurren en un intervalo de tiempo fijo o en un área espacial fija, bajo la suposición de que estos eventos ocurren con una tasa promedio constante y de manera independiente entre sí. Esta distribución es adecuada para contar eventos raros o poco frecuentes.

La función de probabilidad de un modelo de Poisson se define como:

$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

donde:

$ P(X = k)$ es la probabilidad de que ocurra exactamente ( k ) eventos en un intervalo dado.\
$ \lambda $ es la tasa promedio de ocurrencia de eventos (es decir, el número esperado de eventos en el intervalo).\
( k ) es el número de eventos observados (que puede ser 0, 1, 2, ...).\
( e ) es la base de los logaritmos naturales (aproximadamente 2.71828).\

**Ejercicio Resuelto Paso a Paso**

Supongamos que en una sala de cine se observa que en promedio 3 personas llegan en un minuto. Vamos a calcular la probabilidad de que exactamente 4 personas lleguen en un minuto.

**Paso 1: Identificar $\lambda $ y $ k $.**

Aquí, la tasa promedio de llegada $\lambda $ es 3, y queremos calcular la probabilidad de que ( k = 4 ).

**Paso 2: Aplicar la fórmula de la distribución de Poisson.**

Calculamos la probabilidad ( P(X = 4) ) usando la fórmula:

$$
P(X = 4) = \frac{3^4 e^{-3}}{4!}
$$

A continuación la formulación del problema y un histograma que representa la probabilidad para cada $k$.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Definir λ (lambda) y k
lambda_value = 3  # tasa promedio de llegada
k = 4  # número de eventos que queremos calcular

# Paso 2: Calcular P(X = k) usando la fórmula de Poisson
probabilidad = (lambda_value**k) * (math.exp(-lambda_value)) / math.factorial(k)

print(f'La probabilidad de que exactamente {k} personas lleguen en un minuto es: {probabilidad:.4f}')

# Paso 3: Dibujo del histograma de la distribución de Poisson

# Definimos un rango para el número de eventos (k)
k_values = np.arange(0, 11)
probabilities = [(lambda_value**k) * (math.exp(-lambda_value)) / math.factorial(k) for k in k_values]

# Graficar el histograma
plt.bar(k_values, probabilities, color='blue', alpha=0.7)
plt.title('Distribución de Poisson ($\lambda = 3$)')
plt.xlabel('Número de personas que llegan (k)')
plt.ylabel('Probabilidad')
plt.xticks(k_values)
plt.grid(axis='y')

# Mostrar el gráfico
plt.show()
```
