# Regresión Lineal en el Ámbito de la Salud

La regresión lineal es una herramienta estadística fundamental en el ámbito de la salud, utilizada para analizar y predecir la relación entre variables. A continuación, se presentan los puntos clave sobre su aplicación en este campo.

## 1. Definición de Regresión Lineal

La regresión lineal busca modelar la relación entre una variable dependiente $Y$ y una o más variables independientes $X$. La forma general de la ecuación es:

$$
Y = a + bX + \epsilon
$$

donde:
- $Y$: variable dependiente (por ejemplo, presión arterial).
- $X$: variable independiente (por ejemplo, edad).
- $a$: intercepto en el eje $Y$.
- $b$: pendiente de la recta, que indica el cambio en $Y$ por cada unidad de $X$.
- $\epsilon$: término de error que representa la variación no explicada.

## 2. Aplicaciones en Salud

La regresión lineal se emplea en diversas áreas de la salud, tales como:

- **Epidemiología**: para analizar la relación entre factores de riesgo y la incidencia de enfermedades (por ejemplo, indexar el impacto del tabaquismo en la mortalidad).
- **Investigación Clínica**: para evaluar la efectividad de tratamientos (por ejemplo, cómo la dosis de un medicamento afecta los síntomas de una enfermedad).
- **Salud Pública**: para predecir tendencias en datos de salud y tomar decisiones informadas.

## 3. Interpretación de Resultados

Los resultados de un análisis de regresión lineal permiten entender mejor la relación entre variables:

- **Pendiente ($b$)**: indica la dirección y la fuerza de la relación. Un $b$ positivo sugiere que a medida que $X$ aumenta, $Y$ también aumenta; un $b$ negativo indica lo contrario.
- **Intercepto ($a$)**: representa el valor esperado de $Y$ cuando $X$ es cero.
- **R² (coeficiente de determinación)**: mide la proporción de la variabilidad en $Y$ que puede ser explicada por $X$. Un R² cercano a 1 indica un buen ajuste del modelo.

## 4. Supuestos de la Regresión Lineal

Para que los resultados sean válidos, se deben cumplir ciertos supuestos:

- **Linealidad**: la relación entre las variables debe ser lineal.
- **Independencia**: las observaciones deben ser independientes unas de otras.
- **Homoscedasticidad**: la varianza de los errores debe ser constante.
- **Normalidad**: los errores deben seguir una distribución normal (especialmente importante para inferencias).

## 5. Limitaciones

A pesar de su utilidad, la regresión lineal tiene limitaciones:

- **Causalidad**: no establece causalidad; solo describe relaciones.
- **Outliers**: los valores atípicos pueden influir significativamente en el modelo.
- **Linealidad**: no es adecuada para relaciones no lineales sin transformaciones adecuadas.

## 6. Conclusión

La regresión lineal es una herramienta poderosa en el análisis de datos de salud que permite descubrir relaciones y hacer predicciones. Sin embargo, es esencial tener en cuenta sus supuestos y limitaciones al interpretar los resultados.

# Regresión Lineal y Correlación Lineal Baja

## Introducción

La regresión lineal es una técnica estadística que se utiliza para modelar la relación entre una variable dependiente y una o más variables independientes. Un paso clave en la regresión lineal es la evaluación de la correlación entre las variables en estudio. Cuando la correlación lineal entre estas variables es muy baja, se pueden esperar ciertas características en el análisis y sus resultados.

## 1. Expectativas de la Regresión Lineal con Correlación Baja

### 1.1. Relación Débil

- **Descripción**: Una correlación baja (cercana a 0) sugiere que existe una relación débil entre las variables.
- **Implicación**: Esto significa que los cambios en la variable independiente no están asociados de manera consistente con cambios en la variable dependiente.

### 1.2. Bajo Poder Predictivo

- **Descripción**: En una regresión lineal con correlación baja, el modelo puede tener un poder predictivo limitado.
- **Implicación**: Puede ser difícil hacer predicciones precisas sobre la variable dependiente a partir de la variable independiente debido a la falta de relación clara.

### 1.3. Coeficientes No Significativos

- **Descripción**: Al ajustar el modelo, los coeficientes de las variables independientes pueden no ser estadísticamente significativos.
- **Implicación**: Esto indica que, en términos estadísticos, no hay suficiente evidencia para afirmar que la variable independiente tenga un efecto sobre la variable dependiente.

### 1.4. Residuals y Homocedasticidad

- **Descripción**: Los residuos, que son las diferencias entre los valores observados y los valores predichos, pueden mostrar patrones aleatorios sin una tendencia clara.
- **Implicación**: Esto podría indicar que el modelo no explica bien la variabilidad en la variable dependiente, lo que puede llevar a problemas de homocedasticidad.

## 2. Diagnóstico del Modelo

En escenarios donde se observa una correlación baja, es importante realizar diagnósticos adicionales para evaluar el ajuste del modelo:

- **Análisis de Residuos**: Evaluar gráficamente los residuos en función de los valores predichos puede ayudar a identificar patrones que sugieren un mal ajuste del modelo.
- **Exploración de Otras Variables**: Puede ser útil explorar otras variables independientes que puedan tener una relación más fuerte con la variable dependiente.

## 3. Consideraciones Adicionales

### 3.1. Relaciones No Lineales

- **Descripción**: La correlación lineal baja no significa que no exista una relación entre las variables; puede haber relaciones no lineales o interacciones más complejas.
- **Implicación**: Considerar transformaciones de variables o modelos de regresión más complejos (por ejemplo, regresión polinómica) puede ser una solución.

### 3.2. Importancia Contextual

- **Descripción**: Es importante contemplar el contexto del estudio.
- **Implicación**: En algunos casos, es posible que la baja correlación no sea un obstáculo. Dependiendo del dominio, otros factores pueden ser relevantes para las decisiones basadas en el modelo.

## Lanzamos el modelo

```python
# 3. Ajustar un modelo de regresión lineal
X = data['dpi']  # Ingreso disponible per cápita como variable independiente
Y = data['sr']   # Tasa de ahorro como variable dependiente

# Agregar una constante para el intercepto
X = sm.add_constant(X)

# Ajustar el modelo
model = sm.OLS(Y, X).fit()

# 4. Resumen del modelo
print(model.summary())
```

- **Descripción**: El estadístico Durbin-Watson (DW) mide la correlación entre los residuos de un modelo de regresión consecutivos. El valor oscila entre 0 y 4.
- **Interpretación**:
  - **DW = 2**: No hay autocorrelación.

- **Hipótesis**:
  - **Hipótesis nula (H0)**: Los residuos son normalmente distribuidos.
  - **Hipótesis alternativa (H1)**: Los residuos no son normalmente distribuidos.

p-valor significativo (generalmente menor a 0.05):

## ANÁLISIS DE LA NORMALIDAD DE LOS RESÍDUOS

 Residuo en la Regresión Lineal

## Introducción

En el contexto de la regresión lineal, el residuo es una medida que representa la diferencia entre el valor observado de la variable dependiente y el valor predicho por el modelo de regresión. Los residuos son fundamentales para evaluar la calidad del modelo y los supuestos subyacentes.

## 1. Definición de Residuo

- **Fórmula**: El residuo ($e_i$) para la observación $i$ se define como:

$$
e_i = y_i - \hat{y_i}
$$

donde:
- $y_i$: es el valor observado de la variable dependiente.
- $\hat{y_i}$: es el valor predicho por el modelo para la variable dependiente.

## 2. Interpretación de los Residuos

- **Residuo Positivo**: Si el residuo es positivo ($e_i > 0$), significa que el modelo subestima el valor observado; es decir, el valor real es mayor que el predicho.
  
- **Residuo Negativo**: Si el residuo es negativo ($e_i < 0$), significa que el modelo sobrestima el valor observado; el valor real es menor que el predicho.

- **Residuo Cero**: Un residuo de cero ($e_i = 0$) indica que el valor observado coincide exactamente con el valor predicho.

## 3. Importancia de los Residuos

Los residuos son importantes por varias razones:

- **Evaluación del Ajuste del Modelo**: Analizar los residuos permite entender cómo bien se ajusta el modelo a los datos. Residuos más pequeños indican un mejor ajuste.

- **Diagnóstico de Supuestos**: Los residuos ayudan a verificar supuestos clave en la regresión, como la normalidad, homocedasticidad (constancia de varianza) y la independencia. Por ejemplo, un gráfico de residuos puede ser utilizado para verificar la aleatoriedad y la distribución de los residuos.

- **Identificación de Valores Atípicos**: Los residuos también pueden ayudar a detectar outliers o valores atípicos, que son puntos de datos que se desvían significativamente del patrón general.

## 4. Ejemplo de Cálculo de Residuos

```python
# Ajustar un modelo de regresión lineal
X = sm.add_constant(data['dpi'])  # Ingreso disponible per cápita como variable independiente
Y = data['sr']  # Tasa de ahorro como variable dependiente
model = sm.OLS(Y, X).fit()

# Obtener los valores ajustados y los residuos
fitted_values = model.fittedvalues
residuals = model.resid

# Crear la gráfica de residuos
plt.figure(figsize=(10, 6))
plt.scatter(fitted_values, residuals, color='blue', label='Residuos', alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--', label='Media de Residuos')
plt.xlabel('Valores Ajustados')
plt.ylabel('Residuos')
plt.title('Gráfica de Residuos vs. Valores Ajustados')
plt.legend()
plt.grid()
plt.show()
```

```python
# 5. Verificación de supuestos

# 5.1 Normalidad de los residuos
residuos = model.resid
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(residuos, kde=True)
plt.title('Histograma de los Residuos')

plt.subplot(1, 2, 2)
stats.probplot(residuos, dist="norm", plot=plt)
plt.title('Q-Q Plot de los Residuos')

plt.show()
```

## 1. ¿Qué Muestra una Gráfica Q-Q?

- **Ejes**:
  - El eje X representa los cuantiles de la distribución teórica (por ejemplo, la distribución normal).
  - El eje Y representa los cuantiles empíricos de los datos observados.

- **Puntos**: Cada punto en la gráfica Q-Q representa un par de cuantiles, uno de la distribución teórica y otro de la muestra de datos.

## 2. Interpretación de la Gráfica Q-Q

- **Puntos en la Diagonal**:
  - Si los puntos caen en una línea diagonal (recta) que va desde la esquina inferior izquierda a la esquina superior derecha, esto indica que la distribución de los datos observados se asemeja a la distribución teórica.

- **Desviaciones**:
  - **Por encima de la diagonal**: Si los puntos están por encima de la línea diagonal, indica que los datos tienen colas más pesadas que la distribución teórica (es decir, hay más valores extremos).
  - **Por debajo de la diagonal**: Si los puntos están por debajo de la línea diagonal, indica que los datos tienen colas más ligeras que la distribución teórica.

- **Curvaturas**: Las curvas en la gráfica pueden indicar que los datos tienen una distribución diferente a la esperada, sugiriendo que podría ser útil considerar otra distribución.

## 3. Uso en Modelos de Regresión

En el contexto de modelos de regresión, la gráfica Q-Q se utiliza frecuentemente para evaluar el supuesto de normalidad de los residuos. Un residuo normalmente distribuido es uno de los supuestos clave para validar los resultados del modelo de regresión.

# Prueba de Breusch-Pagan

## Introducción

La prueba de Breusch-Pagan es una técnica estadística utilizada para detectar la heterocedasticidad en los residuos de un modelo de regresión lineal. La heterocedasticidad se refiere a la situación en la que la varianza de los residuos no es constante a lo largo de las observaciones, lo que puede afectar la validez de los resultados del modelo.

## 1. Hipótesis de la Prueba

- **Hipótesis nula (H0)**: Los residuos son homocedásticos (la varianza de los residuos es constante).
- **Hipótesis alternativa (H1)**: Los residuos son heterocedásticos (la varianza de los residuos cambia).

## 2. Cálculo del Estadístico de Breusch-Pagan

La prueba de Breusch-Pagan se basa en los residuos del modelo de regresión y calcula un estadístico que se distribuye según una chi-cuadrado (\(\chi^2\)), que se obtiene a partir de la regresión de los residuos al cuadrado sobre las variables independientes del modelo.

El estadístico de prueba es:

$$
BP = \frac{nR^2}{2}
$$

donde:
- \(n\) es el número de observaciones.
- \(R^2\) es el coeficiente de determinación de la regresión de los residuos al cuadrado.

## 3. Resultados y su Interpretación

### 3.1. Estadístico y p-valor

Al realizar la prueba, se obtiene un valor del estadístico de Breusch-Pagan y un p-valor correspondiente. La interpretación de estos valores es la siguiente:

- **p-valor bajo (generalmente < 0.05)**:
  - **Conclusión**: Se rechaza la hipótesis nula. Esto sugiere que hay evidencia significativa de heterocedasticidad, lo que indica que la varianza de los residuos no es constante y puede afectar la validez del modelo.
  
- **p-valor alto (generalmente ≥ 0.05)**:
  - **Conclusión**: No se rechaza la hipótesis nula. Esto sugiere que no hay evidencia suficiente para concluir que hay heterocedasticidad, lo que implica que la varianza de los residuos es constante (homocedástica).

### 3.2. Limitaciones

- **Sensibilidad de la Prueba**: La prueba puede ser sensible a la especificación del modelo. Si el modelo tiene problemas de especificación, pueden producirse resultados engañosos.
- **Falsos Positivos**: En algunas situaciones, el rechazo de la hipótesis nula puede ser debido a la presencia de otros problemas en el modelo, como la multicolinealidad.

## 4. Diagnóstico de Heterocedasticidad

Si se encuentra heterocedasticidad, es importante considerar soluciones como:

- **Transformaciones de Variables**: Aplicar transformaciones logarítmicas o cuadráticas a las variables.
- **Modelos Robustecidos**: Utilizar estimadores de varianza robusta que ajusten los errores estándar a la heterocedasticidad.
