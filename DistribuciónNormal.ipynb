# La ilustración de la distribución normal.

con la función
$$f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parámetros de la distribución normal
mu = 0      # Media
sigma = 1   # Desviación estándar

# Crear un rango de valores x
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calcular la función de densidad de probabilidad (PDF)
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Configurar la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Distribución Normal', color='blue')

# Añadir líneas para la media y las desviaciones estándar
plt.axvline(mu, color='red', linestyle='--', label=f'Media (μ = {mu})')
plt.axvline(mu - sigma, color='green', linestyle='--', label=f'1 σ (μ - σ)')
plt.axvline(mu + sigma, color='green', linestyle='--', label=f'1 σ (μ + σ)')
plt.axvline(mu - 2*sigma, color='orange', linestyle='--', label=f'2 σ (μ - 2σ)')
plt.axvline(mu + 2*sigma, color='orange', linestyle='--', label=f'2 σ (μ + 2σ)')
plt.axvline(mu - 3*sigma, color='purple', linestyle='--', label=f'3 σ (μ - 3σ)')
plt.axvline(mu + 3*sigma, color='purple', linestyle='--', label=f'3 σ (μ + 3σ)')

# Etiquetas y título
plt.title('Distribución Normal')
plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()
```

**Ejercicio: Altura de Pacientes**\
Contexto:

En un hospital, se ha determinado que la altura de los pacientes adultos sigue una distribución normal con una media de 170 cm y una desviación estándar de 10 cm.

**Objetivo:**

Responder a las siguientes preguntas:

¿Cuál es la probabilidad de que un paciente tenga una altura menor a 160 cm?\
¿Qué proporción de pacientes tiene una altura entre 160 cm y 180 cm?\
¿Cuántos pacientes de un total de 1000 se espera que tengan una altura mayor a 185 cm?\

**Solución Paso a Paso**

Paso 1: Importar las Bibliotecas Necesarias\
Primero, importaremos las bibliotecas necesarias para realizar los cálculos y graficar.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

**Paso 2: Definir Parámetros de la Distribución Normal**

Definimos la media y la desviación estándar.

```python
# Parámetros de la distribución normal
mu = 170  # Media en cm
sigma = 10  # Desviación estándar en cm
```

**Paso 3: Calcular la Probabilidad de que un Paciente tenga Menos de 160 cm**

Utilizaremos la función de distribución acumulativa (CDF) para encontrar la probabilidad de que la altura sea menor a 160 cm.

```python
# Probabilidad de que un paciente tenga altura menor a 160 cm
prob_menor_160 = stats.norm.cdf(160, mu, sigma)   # donde stats.norm.cdf es la función, y los valores que definimos
print(f'La probabilidad de que un paciente tenga una altura menor a 160 cm es: {prob_menor_160:.4f}')
```

**Paso 4: Calcular la Proporción de Pacientes con Altura entre 160 cm y 180 cm**

Usaremos la CDF para calcular la probabilidad de que la altura esté entre 160 cm y 180 cm.

```python
# Probabilidad de que un paciente tenga altura entre 160 cm y 180 cm
prob_160_a_180 = stats.norm.cdf(180, mu, sigma) - stats.norm.cdf(160, mu, sigma) # OJO que se resta
print(f'La proporción de pacientes con altura entre 160 cm y 180 cm es: {prob_160_a_180:.4f}')
```

**Paso 5: Calcular el Número Esperado de Pacientes con Altura Mayor a 185 cm**

Calcularemos la probabilidad de que un paciente tenga una altura mayor a 185 cm.

```python
# Probabilidad de que un paciente tenga altura mayor a 185 cm
prob_mayor_185 = 1 - stats.norm.cdf(185, mu, sigma)
# Calcular el número esperado de pacientes en un total de 1000
total_pacientes = 1000
esperados_mayor_185 = prob_mayor_185 * total_pacientes
print(f'Número esperado de pacientes con altura mayor a 185 cm en 1000 pacientes: {esperados_mayor_185:.0f}')
```

**Paso 6: Graficar la Distribución Normal**

Finalmente, graficamos la distribución normal para visualizar mejor la información.

Lo que se pretende en la gráfica es observar las soluciones anteriore mediante las áreas de la función, y se visualiza con los colores.

Se deja como ejercicio interpretar el siguiente gráfico y asociarlo a casa paso.

```python
# Crear un rango de valores x
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = stats.norm.pdf(x, mu, sigma)

# Configurar la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Distribución Normal', color='blue')
plt.fill_between(x, pdf, where=(x < 160), color='red', alpha=0.5, label='Altura < 160 cm')
plt.fill_between(x, pdf, where=((x >= 160) & (x <= 180)), color='green', alpha=0.5, label='160 cm ≤ Altura ≤ 180 cm')
plt.fill_between(x, pdf, where=(x > 185), color='orange', alpha=0.5, label='Altura > 185 cm')

# Añadir etiquetas y título
plt.title('Distribución Normal de Alturas de Pacientes')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad de Probabilidad')
plt.axvline(mu, color='black', linestyle='--', label='Media (170 cm)')
plt.legend()
plt.grid()

# Mostrar la gráfica
plt.show()
```
